```python
from bs4 import BeautifulSoup
import requests
import csv

import pandas as pd
import re
```

```python
url = 'https://www.imdb.com/chart/top/'
```

```python
html_text = requests.get(url).text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
```

```python
names = soup.find_all('td', class_ = 'titleColumn')
names
```

```python
names[0].a.text
```

```python
re.search('\d+', names[0].span.text)[0]
```

```python
ratings = soup.find_all('td', class_ = 'ratingColumn imdbRating')
ratings
```

```python
ratings[0].text.strip()
```

```python
title_list = []
year_list = []
rating_list = []

for i in range(len(names)):
    title_list.append(names[i].a.text)
    year_list.append(int(re.search('\d+', names[i].span.text)[0]))
    rating_list.append(float(ratings[i].text.strip()))
    
df = pd.DataFrame({'Title':title_list, 'Year':year_list, 'Rating':rating_list})
df
```

```python
df.to_csv('Top_250_movies.csv', index = False)
```
