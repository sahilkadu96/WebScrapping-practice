```python
from bs4 import BeautifulSoup
import requests
import csv

import pandas as pd
import re
```

```python
url = 'https://www.indiatimes.com/trending/social-relevance/top-10-richest-cities-in-india-2021-552132.html'
```

```python
html_text = requests.get(url).text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
soup
```

```python
cities = soup.find_all('div', class_ = 'gallery-slide-caption')
cities
```

```python
re.search('\d+\.?', cities[0].text)
```

```python
city_list = []
GDP_list = []

for i in cities:
    city = i.text.split('.')[1].split('-')[0]
    gdp = i.text.split('$')[1].split(' ')[0]
    city_list.append(city)
    GDP_list.append(float(gdp))
```

```python
desc = soup.find_all('div', class_ = 'gallery-slide-description short')
desc
```

```python
description = []

for i in desc:
    description.append(i.p.text)
```

```python
df = pd.DataFrame({'City':city_list, 'GDP in $ Billion':GDP_list, 'Description':description})
df
```

```python
df.to_csv('Top_10_richest_cities_in_India.csv', index = False)
```
